```python
import requests
from eli5 import show_weights, show_prediction
from soft404 import Soft404Classifier
from soft404.utils import html_to_item, item_to_text, item_numeric_features

clf = Soft404Classifier()
```

```python
show_weights(clf.text_clf, vec=clf.vect, top=(30, 30))
```

```python
clf_feature_names = ['text_prob', 'text_len', 'blocks_len', 'max_block_len', 'med_block_len', '0.8_block_len']
show_weights(clf.clf, feature_names=clf_feature_names)
```

```python
def explain_prediction(url):
    html = requests.get(url).text
    print('Prediction for {}: {:.3f}'.format(url, clf.predict(html)))
    text = item_to_text(html_to_item(html))
    return show_prediction(clf.text_clf, text, vec=clf.vect) 
```

```python
explain_prediction('http://orlando.backpage.com/online/classifieds/AdNotFound')
```
