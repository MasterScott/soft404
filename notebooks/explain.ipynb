```python
import copy
import requests
from eli5 import show_weights, show_prediction
from soft404 import Soft404Classifier

soft404_clf = Soft404Classifier()
(_, pre1), (_, pre2), (_, orig_vec), (_, post1), (_, clf) = soft404_clf.pipeline.steps

vec = copy.copy(orig_vec)
orig_preprocessor = vec.build_preprocessor()
def preprocessor(doc):
    return orig_preprocessor(pre2.transform(pre1.transform([doc]))[0])
vec.preprocessor = preprocessor

orig_transform = vec.transform
def transform(X):
    return post1.transform(orig_transform(X))
vec.transform = transform
```

```python
show_weights(clf, vec=vec, top=(30, 30))
```

```python
_request_cache = {}
```

```python
def explain_prediction(url):
    try:
        html = _request_cache[url]
    except KeyError:
        html = _request_cache[url] = requests.get(url).text
    print('Prediction for {}: {:.3f}'.format(url, soft404_clf.predict(html)))
    return show_prediction(clf, html, vec=vec) 
```

```python
explain_prediction('http://orlando.backpage.com/online/classifieds/AdNotFound')
```
